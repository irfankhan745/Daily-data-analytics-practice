```python
import pandas as pd
```

```python
courses = pd.read_csv('/content/courses.csv')
students = pd.read_csv('/content/student_course.csv')
nov = pd.read_csv('/content/reg-month1.csv')
dec = pd.read_csv('/content/reg-month2.csv')
regs=pd.concat([nov,dec],keys=['nov','dec']).reset_index()
delivery = pd.read_csv('/content/deliveries.csv')
```

1. find total revenue generated

```python
regs = regs.merge(courses, how='inner', on='course_id')


regs.head()
```

```python
total_revenue = regs.groupby(['course_name','course_id'])['price'].sum()
total_revenue
```

```python
total_mounth_revenue = regs.groupby(['level_0','course_name','course_id'])['price'].sum()
total_mounth_revenue
```

"Using the regs and students DataFrames, find the total number of courses registered and the total revenue generated by each student in November and December. Display the results showing student name, month, total registrations, and total revenue."

```python
merged_df2 = regs.merge(students,on='student_id',how = 'inner')
monthly_summary = merged_df2.groupby(['name','level_0']).agg(total_revenue = ('price','sum')).reset_index()
monthly_summary.head()
```

Using the delivery DataFrame, calculate the total runs scored by each batting_team. Display the top 5 batting teams with the highest total runs.

```python
batting_team_runs = delivery.groupby(['batting_team'],as_index=False)['total_runs'].sum()
top_5_batting_team = batting_team_runs.sort_values(by='total_runs',ascending=False)
top_5_batting_team.head()
```

Using the delivery DataFrame, calculate the average runs conceded per over for each bowler. Identify the top 5 bowlers who have bowled at least 10 overs and have the highest average runs conceded per over (i.e., the most expensive bowlers).

```python
# grouping
filter_by_bowler_name = delivery.groupby(['bowler'],as_index=False).agg(total_runs = ('total_runs','sum'),total_balls = ('ball','count'))

# overs
filter_by_bowler_name['overs'] = filter_by_bowler_name['total_balls'] / 6

# filter bowlers with at least 10 overs
filter_by_bowler_name = filter_by_bowler_name[filter_by_bowler_name['overs'] >= 10]

# average runs conceded per over
filter_by_bowler_name['run_avg'] = round(filter_by_bowler_name['total_runs'] / filter_by_bowler_name['overs'],2)

# top 5 most expensive bowlers
filter_by_bowler_name.sort_values(by = 'run_avg',ascending=False)
```
